```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
im = cv2.imread('LinkedIn Pic.jpg')
print(im.shape)
```

```python
im = cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
print(im.shape)
im = cv2.resize(im,(1895,1895))
print(im.shape)
plt.imshow(im)
```

```python
orginal_shape = (1895,1895, 3)
all_pixels = im.reshape(-1,3)
all_pixels.shape
```

```python
dominant_colors = 5
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=dominant_colors)
km.fit(all_pixels)
```

```python
centers = km.cluster_centers_
print(centers)
```

```python
centers = np.array(centers,dtype = 'uint8')
print(centers)
```

```python
i = 1

plt.figure(0, figsize=(5,5))

colors = []

for each_color in centers:
    plt.subplot(1,5,i)
    plt.axis('off')
    i+=1
    
    colors.append(each_color)
    
    a = np.zeros((100,100,3), dtype = 'uint8')
    a[:,:,:] = each_color
#     print(a)
    plt.imshow(a)
    
```

```python
new_im = np.zeros((1895*1895, 3),dtype = 'uint8')
print(new_im.shape)
```

```python
colors
```

```python
km.labels_
```

```python
for ix in range(new_im.shape[0]):
    new_im[ix] = colors[km.labels_[ix]]
    
print(new_im.shape)
    
new_im = new_im.reshape((orginal_shape))
plt.imshow(new_im)
plt.show()
```

```python
high_resolution_im = cv2.resize(new_im,(2895,2895))
```

```python
plt.imshow(high_resolution_im)
```

```python
new_im = cv2.cvtColor(high_resolution_im,cv2.COLOR_RGB2BGR)
cv2.imwrite('img_CV2.jpg', new_im, [int(cv2.IMWRITE_JPEG_QUALITY), 90])
```
